NULL DATA CEANNING

```python
# CSV file in mising data(null data) cleanning
```

```python
data={'name': ['a','b','c','d',np.nan],'marks': [85,np.nan,78,90,np.nan],'city': ["ahm", 'Delhi',np.nan,'Mumbai',np.nan]}
df = pd.DataFrame(data)
df
```

```python
df.isnull() # when data is null show True
```

```python
df.isna().sum()  # column wise sum of null values
```

```python
df.dropna() # drop all rows where data is null
```

```python
df.dropna(thresh=2)  # 2 k 2 karta vadhare value hoy to te rahe baki ni drop
```

```python
df.dropna(how='any') # drop all rows where data is null
```

```python
df.dropna(how='all') # #  je row ma badha null hoy a drop
```

```python
#   PYQ
df.dropna(subset=['city']) # city ma jya null hase a column drop
```

```python
df.dropna(axis=0) # axis = x , delete all rows when data is null
```

```python
df.dropna(axis=1) # axis = y , delete all columns when data is null
```
