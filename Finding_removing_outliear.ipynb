Finding and Removing Outliers

FINDING OUTLIERS

```python
'''
45, 47, 52, 52, 53, 55, 56, 58, 62, 80

Q1=50(25 %) Q3=58(75 %)
IQR = Q3-Q1=6
low=Q1-(1.5*IQR) = 43
high=Q3+(1.5*IQR) = 67

'''
```

```python
import pandas as pd
df=pd.read_csv('auto-mpg.csv')
df
```

```python
Q1 = df['acceleration'].quantile(0.25) # 13.825000000000001
Q3 = df['acceleration'].quantile(0.75) # 17.175
IQR=Q3-Q1
low=Q1-(1.5*IQR)
high=Q3+(1.5*IQR)
```

```python
print("low=",low)
print("high=",high)
```

```python
Find_Outliers = df[(df['acceleration']<low) | (df['acceleration']>high)]
Find_Outliers
```

REMOVING OUTLIERS

```python
Remove_Outliers = df[(df['acceleration'] >= low) & (df['acceleration'] <= high)]
Remove_Outliers
```
